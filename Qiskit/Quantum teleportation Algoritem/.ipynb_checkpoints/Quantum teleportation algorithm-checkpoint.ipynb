```python
# Teleportation
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

```python
circuit = QuantumCircuit(3, 3)
```

```python
%matplotlib inline
```

```python
circuit.draw(output="mpl")
```

```python
# now effec a x opreator on first state
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

## now after barrier we have state q0 is in |0> and we are going teleport that state to q2

```python
# first we should have entangelment
circuit.h(1)
circuit.cx(1, 2)
circuit.draw(output='mpl')
```

```python
circuit.cx(0, 1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0, 1], [0, 1])
circuit.draw(output='mpl')
```

```python
# we need two more gate to done teleportation
circuit.barrier()
circuit.cx(1, 2)
circuit.cz(0, 2)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shot=2000).result()
count = result.get_counts()
plot_histogram(count)
```

```python
circuit.measure(2, 2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shot=2000).result()
count = result.get_counts()
plot_histogram(count)
```
